```python
def findDisappearedNumbers(nums):
    n = len(nums)
    bos = []
    myList=set(nums)
    for i in range(1,n+1):
        if i not in myList:
            bos.append(i)
    return bos
# Example usage:
nums = [4,3,2,7,8,2,3,1]
print(findDisappearedNumbers(nums))  # Output: [5, 6]
```
