```python
c = input("Enter three space-separated floats: ")
floats = list(map(float, c.split()))
print(f"{'Number':<12}{'Square':<12}")
print('-' * 20)
for num in floats:
    print(f"{num:<12}{round(num ** 2, 2):<12}")
```

```python
m = int(input("Enter first integer: "))
n = int(input("Enter second integer: "))
total = m + n
count = 0
while total < 50:
    extra = int(input("Enter another integer: "))
    total += extra
    count += 1
print(f"Final sum: {total}")
print(f"Extra numbers entered: {count}")
```
