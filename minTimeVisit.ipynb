```python
# Leet 1266 - Min Time Visiting All Points
```

```python
def deneme(self, points):
    t = 0
    for i in range(len(points) - 1):
        x1, y1 = points[i]
        x2, y2 = points[i + 1]
        t += max(abs(x1 - x2), abs(y1 - y2))
    return t
```

```python
# example
points = [[1,1],[3,4],[-1,0]]
deneme(None, points)
```
