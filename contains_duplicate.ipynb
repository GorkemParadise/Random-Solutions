```python
# Leet 217

nums = [1, 2, 3, 1]
def contains_duplicate(nums):
    num_set = set()
    for num in nums:
        if num in num_set:
            return True
        num_set.add(num)
    return False

contains_duplicate(nums)
```

```python
def dupli2(nums):
    for i in range(len(nums)):
        for j in range(i + 1, len(nums)):
            if nums[i] == nums[j]:
                return True
    return False

dupli2(nums)
```

```python
def dupli3(nums):
    nums.sort()
    for i in range(1, len(nums)):
        if nums[i] == nums[i - 1]:
            return True
    return False

dupli3(nums)
```

```python
def dupli4(nums):
    return len(nums) != len(set(nums))

dupli4(nums)
```

```python
def dupli5(nums):
    from collections import Counter
    counts = Counter(nums)
    for count in counts.values():
        if count > 1:
            return True
    return False

dupli5(nums)
```

```python
def dupli6(nums):
    seen = {}
    for num in nums:
        if num in seen:
            return True
        seen[num] = True
    return False

dupli6(nums)
```

```python
def dupli7(nums):
    return any(nums.count(num) > 1 for num in nums)

dupli7(nums)
```

```python
def dupli8(nums):
    nums.sort()
    return any(nums[i] == nums[i - 1] for i in range(1, len(nums)))

dupli8(nums)
```

```python
def dupli9(nums):
    return len(set(nums)) < len(nums)

dupli9(nums)
```
