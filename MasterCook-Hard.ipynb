```python
def solve():
    n = int(input())
    morson = list(map(int, input().split()))
    burak = list(map(int, input().split()))
    
    # Function to calculate max wins using greedy matching
    def max_wins_possible(available_dishes, morson_dishes):
        # Sort both for greedy matching
        avail = sorted(available_dishes)
        mors = sorted(morson_dishes)
        
        wins = 0
        j = 0  # pointer for available dishes
        
        for m in mors:
            # Find smallest dish that beats m
            while j < len(avail) and avail[j] <= m:
                j += 1
            if j < len(avail):
                wins += 1
                j += 1
        
        return wins
    
    # Calculate maximum possible wins
    max_total_wins = max_wins_possible(burak, morson)
    
    # Construct lexicographically largest ordering
    result = []
    available = burak[:]
    
    for round_idx in range(n):
        # Try dishes from largest to smallest
        best_dish = None
        for dish in sorted(set(available), reverse=True):
            # Check if using this dish maintains max_total_wins
            remaining = available[:]
            remaining.remove(dish)
            
            # Count current wins
            current_wins = sum(1 for i in range(len(result)) if result[i] > morson[i])
            if dish > morson[round_idx]:
                current_wins += 1
            
            # Check future potential
            future_wins = max_wins_possible(remaining, morson[round_idx + 1:])
            
            if current_wins + future_wins >= max_total_wins:
                best_dish = dish
                break
        
        result.append(best_dish)
        available.remove(best_dish)
    
    print(' '.join(map(str, result)))

solve()
```
