```python
'''  You are given an integer array containing positive numbers and zeros. A zero indicates the end of a group, and every sequence of consecutive positive integers between zeros is considered one group. Multiple zeros in a row represent empty groups and should be ignored.Write a C function with the following prototype:int locate_group_start(const int arr[], int size, int goal);The function must:1. Identify each group of positive integers (a group ends when a zero is encountered). 2. Compute the sum of each group. 3. Return the starting index of the first group whose sum equals goal. 4. If no such group exists, return -1.Additional Requirement:
You must correctly check the final group even if the array does not end with a zero.Example:Given the array:{3, 1, 0, 6, 2, 1, 0, 4, 4, 4, 0, 5}Groups are:• Group 1: {3, 1} → sum = 4• Group 2: {6, 2, 1} → sum = 9• Group 3: {4, 4, 4} → sum = 12• Group 4: {5} → sum = 5If goal = 12, the function should return 7 (the starting index of the group {4, 4, 4}).''' 
```

```python
def locate_group_start(arr, size, goal):
    sum = 0
    index = -1

    for i in range(size):
        if arr[i] > 0:
            if sum == 0:
                index = i
            sum += arr[i]
        else:
            if sum == goal:
                return index
            sum = 0
            index = -1

    if sum == goal:
        return index

    return -1
# Example usage:
arr = [3, 1, 0, 6, 2, 1, 0, 4, 4, 4, 0, 5]
goal = 12
result = locate_group_start(arr, len(arr), goal)
print(result)  # Output: 7
```
