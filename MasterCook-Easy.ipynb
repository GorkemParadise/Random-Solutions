```python
def mastercook_competition(N, M, B):
    M_with_index = sorted([(M[i], i) for i in range(N)] )
    B_sorted = sorted(B)

    result = [0] * N
    left = 0
    right = N - 1

    for m_value, m_index in M_with_index:
        if B_sorted[right] > m_value:
            result[m_index] = B_sorted[right]
            right -= 1
        else:
            result[m_index] = B_sorted[left]
            left += 1

    return result
if __name__ == "__main__":
    N = int(input().strip())
    M = list(map(int, input().strip().split()))
    B = list(map(int, input().strip().split()))

    result = mastercook_competition(N, M, B)
    print(" ".join(map(str, result)))
```
