```python
# Example: If A = [4, 2, 5, 2, 8, 4, 3] and k = 3, then the output of your program should be 4.

def midterm(arr, k):
    for x in arr:
        if x in arr:
            arr.remove(x)
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr[k-1]
```

```python
arr = [4, 2, 5, 2, 8, 4, 3]
midterm(arr, 3)
```

```python
def midterm(arr, k):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    h = []
    for x in arr:
        if not h or x != h[-1]:
            h.append(x)
    return h[k - 1]


if __name__ == "__main__":
    arr = [4, 2, 5, 2, 8, 4, 3]
    k = 3
    print(midterm(arr, k))  # 4
```
