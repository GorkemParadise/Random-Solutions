```python
''' 
TankX is a global high-frequency trading firm and a liquidity provider for digital assets, FX, and fixed income markets. Using state-of-art technology infrastructure, advanced mathematical models, and engineering is essential for their business. TankX specializes in data analysis which requires intense study of data obtained from exchanges.

When analyzing the volatility patterns, we can think of the fetched data as a string 
s
 of length 
n
 where every character represents a volatility id (every letter of 
s
 is an uppercase letter from the latin alphabet). Before conducting the analysis some high-volatility subpatterns are removed to make the analysis healthier. Experts in TankX define a substring to be of high-volatility if it is equal to a pre-determined string 
t
 of length 
m
.

The following algorithm is used to generate new analyzable data each time:

1- Remove every single occurence of the string 
t
, in 
s
 simultaneously. After which, concatenate the resulting fragments caused by the operation.

2- Repeat the first step, until no high-volatile subtring remains.

For example, let the string 
s
 be equal to DEAABCBABCCED , and 
t
 be equal toABC. The following will happen:

s
 = DEAABCBABCCED
s
 = DEA B CED
s
 = DEABCED
s
 = DE ED
s
 = DEED
Your task as an intern at TankX, is to report the length of the final string after all operations are done.
'''
```

```python
n, m = map(int, input().split())
s = input().strip()
t = input().strip()
k = len(s)
while True:
    new_s = s.replace(t, "")
    if len(new_s) == k:
        break
    s = new_s
    k = len(s)
print(len(s))
```
